```python
df_from_each_file = (pd.read_csv(f, index_col=[0]) for f in all_files)
#df_from_each_file.reset_index(drop=True, inplace=True)
us_census   = pd.concat(df_from_each_file)
us_census.reset_index(drop=True, inplace=True)
us_census
```

# 1.5

```python
us_census['Income'].replace({'\$': '', ',': ''}, regex=True, inplace=True)
us_census['Income'] = us_census['Income'].astype(float)
us_census.head()
```

# 1.6

```python
us_census[['Men','Women']] = us_census['GenderPop'].str.split('_', expand=True)
us_census.head()
```

# 1.7

```python
us_census['Men'].replace({'M': ''}, regex=True, inplace=True)
us_census['Men'] = us_census['Men'].astype(float)
us_census['Women'].replace({'F': ''}, regex=True, inplace=True)
us_census['Women'] = pd.to_numeric(us_census['Women'], errors='coerce')
us_census.head()
```
